```python
from scipy import stats
import numpy as np


# Add the target variable to the new DataFrame
new_train_df['target'] = train_df['target']

# Calculate z-scores for the target variable
z_scores = stats.zscore(train_df['target'])

# Calculate absolute z-scores
abs_z_scores = np.abs(z_scores)

# Filter the DataFrame based on z-score threshold (e.g., 3)
filtered_entries = abs_z_scores < 3
new_train_df = new_train_df[filtered_entries]
```

```python
# Convert columns to integer data type
new_train_df['year'] = new_train_df['year'].astype(int)
new_train_df['month'] = new_train_df['month'].astype(int)
new_train_df['day'] = new_train_df['day'].astype(int)
new_train_df['departure_hour'] = new_train_df['departure_hour'].astype(int)
new_train_df['departure_minute'] = new_train_df['departure_minute'].astype(int)
new_train_df['arrival_hour'] = new_train_df['arrival_hour'].astype(int)
new_train_df['arrival_minute'] = new_train_df['arrival_minute'].astype(int)
new_train_df['target'] = new_train_df['target'].astype(int)
```

## Encode Categorical Variables

```python
#import encoders and encode categorical variables
from sklearn.preprocessing import LabelEncoder

#select all variables of datatype - object
cat_vars = new_train_df.select_dtypes(include=object)
num_vars = new_train_df.select_dtypes(exclude=object)

#instantiate encoder
le = LabelEncoder()

#encode variables
cat_vars_dummies = pd.get_dummies(data = cat_vars)
cat_vars_label_encoded = cat_vars.apply(le.fit_transform)
```

```python
cat_vars_dummies = cat_vars_dummies.astype(int)
cat_vars_dummies
```

```python
cat_vars_label_encoded
```

## Scale Numeric Variables



```python
#scale variables using standard scaler
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()

num_vars_scaled = scaler.fit_transform(num_vars, )

num_vars_scaled = pd.DataFrame(data = num_vars_scaled, columns=num_vars.columns)
num_vars_scaled.head()
```

```python
# Merge Dataframes

num_vars_plus_dummies = pd.merge(left = num_vars_scaled, right=cat_vars_dummies, left_index=True, right_index=True)
num_vars_plus_label_encoded = pd.merge(left = num_vars_scaled, right=cat_vars_label_encoded, left_index=True, right_index=True)
```

```python
# Split Columns in X and Y

x_num_vars_plus_dummies = num_vars_plus_dummies.drop(labels = 'target', axis = 1)
y_num_vars_plus_dummies = num_vars_plus_dummies['target']

x_num_vars_plus_label_encoded = num_vars_plus_dummies.drop(labels = 'target', axis = 1)
y_num_vars_plus_label_encoded = num_vars_plus_dummies['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

def train_model(algorithm, x, y):
    """
    Function to split training data and train algorithm
    """
    #spliting data into training and testing dataset
    x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=10, train_size=0.8)
    
    #fit and predict
    algorithm.fit(x_train, y_train)
    algorithm_preds = algorithm.predict(x_test)
    
    mae = mean_absolute_error(algorithm_preds, y_test)
    
    return algorithm, mae
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_dummies, forest_dummies_mae = train_model(RandomForestRegressor(), x_num_vars_plus_dummies, y_num_vars_plus_dummies)
forest_encoded, forest_encoded_mae = train_model(RandomForestRegressor(), x_num_vars_plus_label_encoded, y_num_vars_plus_label_encoded)

print('mean absolute error for data containing dummy variables : ', forest_dummies_mae)
print('mean absolute error for data containing encoded variables : ', forest_encoded_mae)
```

```python
# Hyperparameter tuning of random forest

from sklearn.model_selection import GridSearchCV

#train test split
x_train, x_test, y_train, y_test = train_test_split(
    x_num_vars_plus_label_encoded, y_num_vars_plus_label_encoded, random_state=10, train_size=0.8)

#hyperparameters to tune
param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4],
    'min_samples_split': [8, 10],
    'n_estimators': [300, 400, 500]
}
rf = RandomForestRegressor()

rf_grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, 
                              cv=2, n_jobs=-1, verbose=2)

rf_grid_search.fit(x_train, y_train)

best_params = rf_grid_search.best_params_
```

```python
from sklearn.metrics import mean_squared_error

# Extract the best parameters found by the grid search
best_params = rf_grid_search.best_params_

# Instantiate a new RandomForestRegressor with the best parameters
best_rf_model = RandomForestRegressor(**best_params)

# Train the best model on the training data
best_rf_model.fit(x_train, y_train)

# Make predictions on the test data
test_predictions = best_rf_model.predict(x_test)

# Calculate RMSE on the test data
test_rmse = np.sqrt(mean_squared_error(y_test, test_predictions))

print("Root Mean Squared Error (RMSE) on test data:", test_rmse)
```

```python
from sklearn.metrics import mean_squared_error

# Define the evaluate function
def evaluate(model, X_test, y_test):
    # Make predictions on the test data
    y_pred = model.predict(X_test)
    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    return rmse, y_pred

# Get the best estimator from the grid search
rf_best_grid = rf_grid_search.best_estimator_

# Evaluate the best estimator on the test data
rf_grid_accuracy, predictions = evaluate(rf_best_grid, x_test, y_test)

print("Root Mean Squared Error (RMSE) on test data:", rf_grid_accuracy)
```
